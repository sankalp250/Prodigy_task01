```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('worldpopulationdata.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['Country Name'].nunique()
```

Dropping unnecessary columns

```python
df.drop(['Series Name'],axis=1,inplace=True)
```

```python
df.columns
```

Extraction of top 10 countries wrt population


```python
total_population = df[df['Series Code'] == 'SP.POP.TOTL']

total_population_sorted = total_population.sort_values(by="2022", ascending=False)

top_ten_countries = total_population_sorted.head(10)
print("Top ten countries of total population\n")
print(top_ten_countries[['Country Name']] )
```

Difference between the year 2022 to 2012

```python

plt.figure(figsize=(12, 6))
plt.subplot(2,2,1)
sns.barplot(x="2022", y="Country Name", data=top_ten_countries,orient='h') # type: ignore
plt.title("Top Ten Countries of Total Population (2022)")
plt.xlabel("Total Population")
plt.ylabel("Country")
plt.show()

plt.figure(figsize=(12, 6))
plt.subplot(2,2,2)
sns.barplot(x="2012", y="Country Name", data=top_ten_countries,orient='h') # type: ignore
plt.title("Top Ten Countries with Total Population (2012)")
plt.xlabel("Total Population")
plt.ylabel("Country")
plt.show()
```

Total male population rise in top 10 most populated country

```python
total_male_population = df[df['Series Code'] == 'SP.POP.TOTL.MA.IN']

total_male_pop = total_male_population.sort_values(by="2022",ascending= False)
male_top_countries = total_male_pop.head(10)

print(male_top_countries[['Country Name']])

```

```python
sns.barplot(x = "2022" , y ="Country Name" , data = male_top_countries,orient='h')
plt.title('Top 10 countries with most male population')
plt.xlabel('male population')
plt.ylabel('country')
plt.show()

sns.barplot(x = "2012" , y ="Country Name" , data = male_top_countries,orient='h')
plt.title('Top 10 countries with most male population')
plt.xlabel('male population')
plt.ylabel('country')
plt.show()
```

```python
total_female_population = df[df['Series Code'] == 'SP.POP.TOTL.FE.IN']

total_female_pop = total_female_population.sort_values(by="2022",ascending= False)
female_top_countries = total_female_pop.head(10)

print(female_top_countries[['Country Name']])
```

```python
sns.barplot(x = "2022" , y ="Country Name" , data = female_top_countries,orient='h')
plt.title('Top 10 countries with most female population')
plt.xlabel('female population')
plt.ylabel('country')
plt.show()

sns.barplot(x = "2012" , y ="Country Name" , data = female_top_countries,orient='h')
plt.title('Top 10 countries with most female population')
plt.xlabel('female population')
plt.ylabel('country')
plt.show()
```
